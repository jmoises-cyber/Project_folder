```python
from google.colab import auth
import gspread
from google.auth import default
import pandas as pd

# Authenticate to Google
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)

# Replace with your sheet name and worksheet name
sheet_name = 'Homecore_Credit'
worksheet_name = 'data'

try:

  # Open the spreadsheet and select the worksheet
  sh = gc.open(sheet_name)
  worksheet = sh.worksheet(worksheet_name)
  print("here")
  # Get all values from the worksheet
  data = worksheet.get_all_values()

   # Convert to a pandas DataFrame (assuming the first row is the header)
  df = pd.DataFrame(data[1:], columns=data[0])



  print("Data loaded successfully!")
  #display(df.head())


except Exception as e:
  print(f"An error occurred: {e}")
  print("Please double-check your sheet name and worksheet name, and ensure you have sharing permissions set correctly for the sheet.")
```
